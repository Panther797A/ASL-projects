```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

people = ('Europe', 'Asia', 'Australia', 'Africa', 'USA')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis() 
ax.set_xlabel('Performance')
ax.set_title('Average consumtion of beverage around the globe!!')

plt.show()
```

```python
import matplotlib.pyplot as plt
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

print(color.BOLD + '                Average consumption of COCA-COLA around the globe!' + color.END)

labels = 'North America', 'Global Ventures', 'Latin America', 'Europe, Middle East, Africa', 'Asia Pacific'
sizes = [18, 3, 28, 28, 23]
explode = (0, 0.1, 0, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 



plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['2010', '2012', '2014', '2016', 'Present']
USA = [20, 34, 30, 35, 27]
Europe = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, USA, width, label='USA')
rects2 = ax.bar(x + width/2, Europe, width, label='Europe, Middle East, Africa')

ax.set_ylabel('Scores')
ax.set_title('Average consumption of Coca-Cola between USA and Europe, Middle East, Africa')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```
